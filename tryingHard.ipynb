```python
def swap_rows(A, a, b):
    assert (a >= 0) and (b >= 0)
    N = A.shape[0]
    assert (a < N) and (b < N)
    temp = A[a, :].copy()
    A[a, :] = A[b, :].copy()
    A[b, :] = temp
```

```python
def BackSub1(aug_matrix,x):
    N = x.size
    [Nrow,Ncol] = aug_matrix.shape
    assert Nrow+1 == Ncol
    assert N == Nrow
    for row in range(N-1,-1,-1):
        RHS = aug_matrix[row,N]
        for coloumn in range(row+1,N):
            RHS = RHS - x[coloumn]*aug_matrix[row,coloumn]
        x[row]= RHS/aug_matrix[row,row]
    return
```

```python
def GaussElimination(A,b):
    [Nrow,Ncol] = A.shape
    assert Nrow == Ncol
    N = Nrow
    assert b.size == N
    aug_mat = np.zeros((N,N+1))
    aug_mat[0:N,0:N] = A
    aug_mat[:,N] = b
    for coloumn in range(0,N):
        for row in range(coloumn+1,N):
            mod_row = aug_mat[row,:]
            mod_row -= (mod_row[coloumn]/aug_mat[coloumn,coloumn]*aug_mat[coloumn,:])
        aug_mat[row] = mod_row
    return aug_mat
```
